## Comparing Apple and Samsung Stock Performance

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
apple_stock = yf.download("AAPL", start = "2022-01-01", end = "2023-01-01")
samsung_stock = yf.download("SSNLF", start = "2022-01-01", end = "2023-01-01")
```

## Daily Return Calculation and Visualization

#### Change of stock price over a day

```python
apple_stock["Daily_Return"] = apple_stock["Adj Close"].pct_change()
samsung_stock["Daily_Return"] = samsung_stock["Adj Close"].pct_change() 
```

```python
apple_stock
```

```python
samsung_stock
```

```python
plt.plot(apple_stock["Daily_Return"], color='orange', label="Apple")  
plt.plot(samsung_stock["Daily_Return"], color='blue', label="Samsung")  

plt.title("Apple vs Samsung Returns")
plt.xlabel("Date")
plt.ylabel("Daily Return")
plt.legend() 
plt.show()
```

## Calculation Simple Moving Averages

```python
apple_stock["50_MA"] = apple_stock["Adj Close"].rolling(window = 50).mean()
samsung_stock["50_MA"] = samsung_stock["Adj Close"].rolling(window = 50).mean()
```

```python

# Creating subplots with 2 rows and 1 column
fig, axs = plt.subplots(2, 1, figsize=(10, 12))

# First subplot
axs[0].plot(apple_stock["Adj Close"], label="Adj Close", color='green')
axs[0].plot(apple_stock["50_MA"], label = "50-day MA", color='orange')
axs[0].set_title("Apple Stock Simple Moving Average")
axs[0].set_xlabel("Date")
axs[0].set_ylabel("Price ($)")
axs[0].legend()

# Second subplot
axs[1].plot(samsung_stock["Adj Close"], label="Adj Close", color='blue')
axs[1].plot(samsung_stock["50_MA"], label = "50-day MA", color='orange')
axs[1].set_title("Samsung Stock Simple Moving Average")
axs[1].set_xlabel("Date")
axs[1].set_ylabel("Price ($)")
axs[1].legend()

# Adjust layout to prevent overlapping of subplots

# Show the plots
plt.show()
```

## Stock Performance Analysis: Apple vs. Samsung

```python
apple_highest = apple_stock["Close"].max()
apple_lowest = apple_stock["Close"].min()
samsung_highest = samsung_stock["Close"].max()
samsung_lowest = samsung_stock["Close"].min()

# Calculate percentage returns (using closing prices)
apple_new_return = ((apple_highest - apple_lowest) / apple_lowest) * 100
samsung_new_return = ((samsung_highest - samsung_lowest) / samsung_lowest) * 100

# Print the results
print("Apple Highest Price:", apple_highest)
print("Apple Lowest Price:", apple_lowest)
print("Apple Return:", apple_new_return, "%")

print("\nSamsung Highest Price:", samsung_highest)
print("Samsung Loweest Price:", samsung_lowest)
print("Samsung Return:", samsung_new_return, "%")


if apple_new_return > samsung_new_return:
    best_stock = "Apple"
else:
    best_stock = "Samsung"
print("\nThe best performing stock with the highest return is:", best_stock)
```
